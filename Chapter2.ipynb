## 4. The KEGG annotation of genes  

- KEGG(京都基因与基因组百科全书)是基因组破译方面的数据库。在后基因时代一个重大挑战是如何使细胞和有机体在计算机上完整的表达和演绎，让计算机利用基因信息对更高层次和更复杂细胞活动和生物体行为作出计算推测。为达到此目的，人们建立了一个在相关知识基础上的网络推测计算工具。在给出染色体中一套完整的基因的情况下，它可以对蛋白质交互（互动）网络在各种细胞活动起的作用作出预测。 KEGG 的**PATHWAY** 数据库整合当前在分子互动网络（比如通道，联合体）的知识，KEGG 的GENES/SSDB/KO 数据库提供关于在基因组计划中发现的基因和蛋白质的相关知识，KEGG 的COMPOUND/GLYCAN/REACTION数据库提供生化复合物及反应方面的知识。

## 5. The GO enrichment of genes 基因富集

- 一旦我们有了一组所需格式的基因/蛋白质，通过功能丰富(如GO terms)来注释它们在生物学中的作用是非常有趣的。富集的目的是为给定的基因找到一组重要的功能(这里，我们在途径富集的情况下使用GO术语或pathway)。富集是通过在基因组中寻找GO术语的过度表达来实现的。过度代表性的显著性是通过统计检验如Fisher检验来计算的。

（1）Install and load the topGO library and the ALL dataset for GO enrichment analysis

```r
library(topGO)
library(ALL)
data(ALL)
data(geneList)
```

(2)为affymetirx芯片数据设置注释，注释来自ALL数据集

```r
affyLib <- paste(annotation(ALL), "db", sep=".")
library(package=affyLib, character.only=TRUE)
```

(3) 从GO富集基因库，使用genelist,可以检查得到p值<0.01的基因，这些基因为差异表达基因

```r
 sum(topDiffGenes(geneList))  
# should come as 50
```

（4）当输入数据和库准备就绪时，我们必须首先创建一个topGOdata
对象

```r
myGOData <- new("topGOdata", ontology="BP", allGenes=geneList, geneSel=topDiffGenes, nodeSize=10, annot= annFUN.db, 
affyLib=affyLib)
```

```r
myGOData
```

（5）富集检验

- fisher

```r
 Myenrichment_Fisher <- runTest(myGOData, algorithm= "classic", statistic="fisher")
```

```r
 score(Myenrichment_Fisher) ##查看富集分数
```

- KS 检验

```r
Myenrichment_KS <- runTest(myGOData, algorithm= "classic", 
statistic="ks")
```

（6）我们可以通过以下表格的形式看到这些测试的结果
GenTable功能:

```r
enrich_table <- GenTable(myGOData, classicFisher=Myenrichment_Fisher,topNodes = 20)

head(enrich_table)
```

> topGOdata对象由基因、它们的得分、GO注释和GO层次结构信息构成一个嵌套列表。
所有的富集分析都是对这些数据进行的。
要创建topGOdata对象，您需要一些其他输入，如nodeSize来修剪GO树(GO有一个层次树结构)和一个库来映射GO的基因。
一旦我们有了目标，我们就开始进行浓缩测试。在我们的例子中，我们使用Fisher测试来发现丰富的基因。
runTest函数的输入是一个topGOdata对象，以及测试统计信息(这里是Fisher测试)和测试算法(在我们的示例中是经典算法)。
该函数返回一个topGOresult对象，如下所示:

- 您可以通过绘制p值来观察彼此之间的测试性能
反对对方。为此，我们需要提取GO项和相应的
的termStat函数中包含的基因数量和其他信息
下面的命令:

```r
gostat <- termStat(myGOData, names(score(Myenrichment_Fisher)))

plot(score(Myenrichment_Fisher), score(Myenrichment_KS)
[names(score(Myenrichment_Fisher))], xlab="P values Fisher test", 
ylab="P values KS test", cex=(gostat$Annotated/max(gostat$Annotated))*4, 
col=heat.colors(gostat$Significant))
```

 - 散点图显示了富集之间的比较
两种不同的基因检测方法即Fisher检验和KS检验。散点图显示
